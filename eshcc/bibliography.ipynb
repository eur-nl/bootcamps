### Selected bibliography

#### Programming Commandline

Jeroen Janssens, Data Science at the Command Line, O'Reilly , 2014: https://www.datascienceatthecommandline.com/
(Uses the commandline to analyse data, using csv files, sqlite, etc.

#### Programming (Python)
There are many excellent tutorials and books about programming with Python. You might want to try:

Zed Shaw, Learn Python The Hard Way: https://www.learnpythonthehardway.org

#### Programming (R)
And the same goes for programming in R, some very good books around. You might want to try:

Robert I. Kabacoff, R in Action (second edition). Data analysis and graphics with R, 2015 (Manning): https://www.manning.com/books/r-in-action-second-edition

#### Download Anaconda Distribution
https://www.anaconda.com/download/

#### Jupyter Notebook
http://jupyter.org (Part of the Anaconda distribution) as is the new Jupyter Lab notebook.

#### R programming environment
Download the R language for your OS at: https://www.r-project.org
Download R Studio for your OS: https://www.rstudio.com/products/RStudio/

#### Web scraping (Python)
Beautifulsoup. Python library often used for web scraping: https://www.crummy.com/software/BeautifulSoup/bs4/doc/

Another one is Scrapy: https://scrapy.org/

Jasper also mentioned lxml: A nice lxml introduction can be found in The Hitchhiker's Guide to Python: http://docs.python-guide.org/en/latest/scenarios/scrape/

#### Web scraping (R)
We have never used R for webscraping ourselves, but it seems that the R "rvest" library (inspired by Beautifulsoup) is used quite a lot: https://github.com/hadley/rvest

#### Data, statistics, and computing (Python programming language)
pandas: powerful Python data analysis toolkit documentation, at: http://pandas.pydata.org/pandas-docs/stable/index.html

Jake VanderPlas, Python Data Science Handbook. Essential Tools for Working with Data, O'Reilly, 2016, freely available at: https://jakevdp.github.io/PythonDataScienceHandbook/ (With a nice chapter on the topic of "feature engineering": https://jakevdp.github.io/PythonDataScienceHandbook/05.04-feature-engineering.html)

Wes McKinney, Python for Data Analysis. Data Wrangling with Pandas, NumPy and IPython, O'Reilly, 2017 (2nd edition; 1st edition 2012 and downloadable from various websites). The second edition of this important book in the context of data analysis and Python is updated for Python 3.6 and provides information about using Python, IPython and the Jupyter Notebook as well as the important Python scientific libraries: NumPy, Matplotlib, Pandas, SciPy, Statslib, and Sckit-Learn.

Alice Zheng and Amanda Casari, Feature Engineering for Machine Learning. Principles and Techniques for Data Scientists, O'Reilly (March 2018): http://shop.oreilly.com/product/0636920049081.do (All examples in Python; stack used: Numpy, Pandas, Scikit-learn, Matplotlib, and Seaborn).

#### Statistics and computing (R programming language)
Chester Ismay and Albert Y. Kim, ModernDive. An Introduction to Statistical and Data Sciences via R, 2017, online publication, retrieved on 2017-11-08 at: http://moderndive.com/index.html

Gareth James, Daniela Witten, Trevor Hastie, and Robert Tibshirani, An Introduction to Statistical Learning with Applications in R, New York, Springer Science and Business Media, 2013, http://www-bcf.usc.edu/~gareth/ISL/index.html Python code for some chapters of the book can be found here: https://github.com/JWarmenhoven/ISLR-python)

Trevor Hastie, Robert Tibshirani, and J. Friedman, Elements of Statistical Learning, 2nd ed., New York, Springer Science and Business Media, 2009, http://statweb.stanford.edu/~tibs/ElemStatLearn/

#### Tidy Data (Hadley Wickham, R programming language)
Hadley Wickham, Tidy Data, in: Journal of Statistical Software, August 2014, Vol. 59, issue 10, https://www.jstatsoft.org/article/view/v059i10/v59i10.pdf. Supplementary material (datasets and R code) available on GitHub: https://github.com/hadley/tidy-data.

Hadley Wickham and Garrett Grolemund, R for Data Science. Visualize, Model, Transform, and Import Data, O'Reilly, 2017. Online version freely available: http://r4ds.had.co.nz/

Tidyverse R library, retrieved from: https://www.tidyverse.org/ [R packages for data science]

#### Tidy Data (Python programming language)
Jean-Nicholas Hould, Tidy Data in Python (06/12/2016) published on Jean-Nicholas Hould on Data Science: http://www.jeannicholashould.com/tidy-data-in-python.html Which is based on Hadley Wickham's article "Tidy Data" (see above).

#### Text analysis (Python)
NLTK (from their website: NLTK has been called “a wonderful tool for teaching, and working in, computational linguistics using Python,” and “an amazing library to play with natural language.”). Large library, but well documented, start at: https://www.nltk.org/

Spacy. More or less "the new kid on the block" in the Python eco-system, claims easy connections to Machine Learning libraries: https://spacy.io/

#### Text analysis (R)
TidyText. Julia Silge and David Robinson wrote: Text mining with R; A tidy approach (O'Reilly, 2018), freely available at: https://www.tidytextmining.com/

TM is a R library for text mining: https://cran.r-project.org/web/packages/tm/vignettes/tm.pdf

Matthew Jockers, Text Analysis with R for Students of Literature (Springer, Quantitative Methods in the Humanities and Social Sciences, 2014; behind paywall), more info: http://www.springer.com/us/book/9783319031637
