Use a for loop to check if a word is a [palindrome](https://en.wikipedia.org/wiki/Palindrome)  
Ignore any white spaces

word1 = "Never odd or even"  
word2 = "Not a palindrome"

````python
word1 = "Never odd or even"
word2 = "Not a palindrome"

# Your code here
````

````python
word1 = "never odd or even"
word2 = "not a palindrome"

word1 = word1.replace(" ", "")
word2 = word2.replace(" ", "")

is_palindrome = True
for i in range(len(word1) // 2 + 1):
    is_palindrome &= word1[i] == word1[-1 - i]
print(is_palindrome)

is_palindrome = True
for i in range(len(word2) // 2 + 1):
    is_palindrome &= word2[i] == word2[-1 - i]
print(is_palindrome)
````

````python
a = [1, 2, 3, 2, 1]
# Your code here

b = [1, 2, 3, 4, 1]
# Your code here
````

````python
a = [1, 2, 3, 2, 1]
b = [1, 2, 3, 4, 1]

a_r = a[:]
a_r.reverse()
b_r = b[:]
b_r.reverse()

# An even faster way would be 
a_r = a[::-1]
b_r = b[::-1]

print(a == a_r)
print(b == b_r)
````
